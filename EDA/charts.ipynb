```python
import matplotlib.pyplot as plt
import numpy as np          
```

```python
data = [140, 189, 185, 195, 149, 189, 147, 154, 174, 169, 195,
        159, 192, 155, 191, 153, 157, 140, 144, 140, 172, 157,
        181, 182, 166, 167,140,140]
unique_values, counts = np.unique(data, return_counts=True)
# Sturges rule for making bins by calculating the number of bins
k = int(np.ceil(1 + np.log2(len(data))))
plt.figure(figsize=(23, 3))
plt.hist(data, bins=k, edgecolor="yellow")
plt.xticks(unique_values)
plt.show()
```
