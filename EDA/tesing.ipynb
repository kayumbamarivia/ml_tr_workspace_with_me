```python
import requests as req
import pandas as pd
from sklearn.preprocessing import LabelEncoder


```

```python
try:
    users_api = req.get('http://127.0.0.1:8000/auth/users')
    users_api.raise_for_status()
    users_data = users_api.json()
    users_df = pd.DataFrame(users_data)
    
    teams_api = req.get('http://127.0.0.1:8000/auth/teams')
    teams_api.raise_for_status()
    teams_data = teams_api.json()
    teams_df = pd.DataFrame(teams_data)
    
    print(users_df.head())
    print("\n")
    print(teams_df.head())
    print("\n")
    
    merged_df = pd.merge(users_df, teams_df, left_on="id", right_on="user_id", how="inner")
    print("Merged DataFrame (Users + Teams):")
    print(merged_df.head())
    
    # merged_df['name_x'] = merged_df["name_x"].astype(int)
    merged_df['name_x'] = pd.to_numeric(merged_df["name_x"], errors="coerce")
    print("\n",merged_df.dtypes)
    
    print(merged_df.describe())
    
    
    label_encoder = LabelEncoder()

    # Apply One-Hot Encoding (example on a column 'name_x')
    # One-Hot Encoding creates binary columns for categorical values
    if 'name_x' in merged_df.columns:
        one_hot_encoded_df = pd.get_dummies(merged_df, columns=['name_x'])
        print("\nOne-Hot Encoded DataFrame:")
        print(one_hot_encoded_df.head())
    
except req.exceptions.RequestException as e:
    print(f"Error fetching from api: {e}")
    
```
